All views are equal to 100, columns containing this information can be discarded.

```python
df = df[df.columns.drop(list(df.filter(regex='View')))]

df = df.T
df.columns = df.iloc[0]
df = df[1:]
df.reset_index()

df.head()
```

```python
mean = df.mean(axis=0)
mean = mean.to_frame().reset_index()
mean = mean.rename(columns= {0: 'Mean'})
mean['Mean'] = mean['Mean'].astype(float)

mean.head()
```

```python
print("Yes, there is a colors that provide more links than blue color. Let's see at least top two: ")
mean.nlargest(2, 'Mean')
```

Fred doesn't know how to interpret the results correctly and isn't sure
which link follows were purposeful and which were accidental. 

Null hypothesis:
The ultramarine color does not differ from the Navy color in the number of link follows due to the randomness of such follows, so their interchange will not affect the result.
    
Alternative hypothesis:
Ultramarine actually has more purposeful follows.

Let's set a level of significance at α = 0.05

```python
a = 0.05

p_value = pearsonr(df['Navy'], df['Ultramarine']).pvalue # Pearson coefficient
```

```python
if p_value > a:
    print(f'P-value result is: {p_value} and since it bigger that our α = {a} the correlation is not statistically significant.  We fails to reject the null hypothesis')
else:
    print(f'P-value result is: {p_value} and since it lesser that our α = {a} we reject the null hypothesis')
```

As a result, there is a 25% chance that a large number of link follows are random and a 95% chance that the result is a consequence of the created conditions.
